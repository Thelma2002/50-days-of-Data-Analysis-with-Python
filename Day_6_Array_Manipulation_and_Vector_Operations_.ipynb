Create a 2-dimensional array of random integers from 0 to
100 with the shape (5, 5). Use the array to find the
minimum and maximum values, as well as the mean and
standard deviation. Ensure that the results are
reproducible.

```python
rng = np.random.default_rng(seed=42)
array1 = rng.integers(low=0, high=100, size=(5,5))

min_val = array1.min()
max_val = array1.max()
mean_val = array1.mean()
std_val = array1.std()
```
