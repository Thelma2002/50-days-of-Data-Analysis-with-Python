2. Here is a list below:  
list1 = [2, 3, 4, 6]

* Create a one-dimensional array from the list above. Write
another line of code to check the shape and size of your
array.
* Write a function (Python function) that takes this array and
returns a new array with all the values in the original array
reversed.

```python
#Creating a 1-dimensional array from the list
list1=[2,3,4,5]
arr = np.array(list1)

#Check the shape and size of the array
print("Shape:", arr.shape)
print("Size:", arr.size)

#Define a function to reverse the array
def reverse_array(array):
  return array[::-1] #slicing trick to reverse
#Use the function
reverse_arr = reverse_array(arr)
print("Reversed array:", reverse_arr)
```

4. Create an array from the three (3) lists. Take note
that some of the lists contain floats.
Create an array that
will have an int64 data type.  

```python
list1 = [2, 3, 4, 6]
list2 = [8, 10.1, 12, 14]
list3 = [16, 18, 20, 22.1]
#create an array from the lists
arr3 = np.array([list1, list2, list3], dtype=np.int64)
print("Shape:", arr3.shape)
print("Dimensions:", arr3.ndim)
print("Data type:", arr3.dtype)
```
