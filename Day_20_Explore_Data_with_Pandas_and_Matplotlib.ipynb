```python
import pandas as pd
products = ["sugar", "salt", "oil", "diapers", "rice"]
costs = [2450, 1989, 6745, 9807, 8743]
sales = [27908, 4508, 6734, 9976, 9000]
```

1. Create a pandas DataFrame from the lists above. Find the
most profitable product. Using Matplotlib, create a bar
plot to visualize product profitability.

```python
df = pd.DataFrame({'products': products, 'costs': costs, 'sales': sales})
# Profit for each product
df['profit'] = df['sales'] - df['costs']
# Index of most profitable product
most_profitable_product = df['profit'].idxmax()
# Use index to find the corresponding product
most_profitable_product_name = df.loc[most_profitable_product, 'products']
print(f'The most profitable product is {most_profitable_product_name}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.bar(df['products'], df['profit'])
plt.xlabel('Products')
plt.ylabel('Profit')
plt.title('Profitability of product', fontsize=20)
plt.show()
```

2. Using just pandas, find the least profitable product.

```python
# Find the least profitable product
least_profitable_index = df['profit'].idxmin()
least_profitable_product = df.loc[least_profitable_index, 'products']
print(f'The least profitable product is {least_profitable_product}')
```

3. What is the difference in profit between the most
profitable item and the least profitable item?

```python
# calculate most profitable in profit
most_profitable = df.loc[most_profitable_product, 'profit']
# calculate least profitable in profit
least_profitable = df.loc[least_profitable_index, 'profit']
# Calculate the difference
difference = most_profitable - least_profitable
print(f'The difference in profit between the most profitable item and the least profitable item is {difference}')
```

4. Using pandas, and Matplotlib, create a line plot of the
costs and profits of all products.

```python
df.plot(x='products', y=['costs', 'profit'], kind='line', figsize=(12,8))
plt.xlabel('Products')
plt.ylabel('Costs and Profit')
plt.title("Cost and profit analysis", fontsize=20)
plt.show()
```

5. Using pandas, calculate the average cost per product.

```python
average_cost = df['costs'].mean()
print(f'The average cost per product is {average_cost}')
```

6. Using pandas calculate the average profit per product.

```python
average_profit = df['profit'].mean()
print(f'The average profit per product is {average_profit}')
```
