Your task is to import the CSV file using NumPyâ€™s
genfromtxt() function.
 1. Transpose the array.
 2. Using slicing, create three arrays: an array of all the
names in the column, an array of the age column,
and an array of the gender (sex) column.

```python
import numpy as np
# Import a csv file
arr = np.genfromtxt('names_age_sex_data.csv', delimiter = ',', dtype = 'str')


#transpose array
transposed_arr = arr.T


# Slice to create three array
names = transposed_arr[0:1, 1:]
age = transposed_arr[2:3, 1:]
gender = transposed_arr[1:2, 1:]
print(names)
print(age)
print(gender)
```

Using NumPy, write code that returns all the names of
people who are 56 years old.

```python
# Return all the names of people who are 56 years
aged_56 = names[age == '56']
print(aged_56)
```

How many people are aged 44 or under?

```python
# Number of people aged 44 or under
age_44_under = names[age <= '44']
num_of_44_under = np.shape(age_44_under)
print(num_of_44_under)
```

Write another code snippet to return an array of all the
males in the dataset. How many males are in the dataset?

```python
# array of all males in the dataset
males = names[gender == 'Male']
num_of_males = np.shape(males)
print(num_of_males)
```

Calculate the average age of all the females in the table.

```python
# Average of all females
age = age.astype(int)
avg_females = np.average(age[gender == 'Female'])
print(avg_females)
```

Calculate the average age of people named Olivia and
Kate.  

```python
# average of people named Olivia
olivia = np.average(age[names == 'Olivia'])
print(olivia)

# average of people named Kate
kate = np.average(age[names == 'Kate'])
print(kate)

# average of both olivia and kate
avg_olivia_kate = (olivia + kate)/2
print(avg_olivia_kate)
```
