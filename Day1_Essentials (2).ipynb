Mathematical functions.

*   np.add() and npp.subtract()
*   np.multiply() and np.divide()
*   np.power() and np.sqrt()






```python
#Addition
np.add(5, 3)   # → 8
np.add([1, 2], [3, 4])   # → [4, 6]

#subtraction
np.subtract(5, 3)   # → 2
np.subtract([10, 20], [1, 2])   # → [9, 18]

#Multiplication
np.multiply(2, 4)   # → 8
np.multiply([1, 2, 3], 2)   # → [2, 4, 6]

#Division
np.divide(10, 2)   # → 5.0
np.divide([10, 20], [2, 5])   # → [5.0, 4.0]

#Power
np.power(2, 3)   # → 8   (2³)
np.power([1, 2, 3], 2)   # → [1, 4, 9]

#square root
np.sqrt(16)   # → 4.0
np.sqrt([1, 4, 9, 16])   # → [1., 2., 3., 4.]
```

Statistical functions.
* np.mean(): It flattens the array and calculates the mean of a flattened array. You can specifiy the axis.
* np.median()
* np.std(): Measure of how spread out the data is. It tells how much the data points typically vary from the average (mean).
* np.var(): measures the average of the squared differences from the mean. Calculates the variance if an array.
* np.min() and np.max()

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
# Mean
np.mean(arr)        # → 3.5   (average of all numbers)
np.mean(arr, axis=0) # → [2.5, 3.5, 4.5] (average column-wise)
np.mean(arr, axis=1) # → [2.0, 5.0]       (average row-wise)

#Median
np.median(arr)        # → 3.5
np.median(arr, axis=0) # → [2.5, 3.5, 4.5]
np.median(arr, axis=1) # → [2.0, 5.0]

#Standard deviation
np.std(arr)        # → 1.707... (overall spread)
np.std(arr, axis=0) # → [1.5, 1.5, 1.5] (spread per column)
np.std(arr, axis=1) # → [0.816..., 0.816...] (spread per row)

#Variance
np.var(arr)        # → 2.916...
np.var(arr, axis=0) # → [2.25, 2.25, 2.25]
np.var(arr, axis=1) # → [0.666..., 0.666...]

#Min and max
np.min(arr)        # → 1
np.min(arr, axis=0) # → [1, 2, 3] (min of each column)
np.min(arr, axis=1) # → [1, 4]    (min of each row)

np.max(arr)        # → 6
np.max(arr, axis=0) # → [4, 5, 6] (max of each column)
np.max(arr, axis=1) # → [3, 6]    (max of each row)
```

Data Cleaning
* dropna(): If we have NaN values in DataFrame we can drop them using the df.dropna() method.
* fillna(): Used to fill missing values in the DataFrame with a specified value or method
* bfill(): Backward fill fills missing values with the next non-missing value in the column.
* ffill(): Forward fill fills the missing values with the previous non_missing value in the column.

```python
# Create a DataFrame with some missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, np.nan, 35, 40],
    "City": ["New York", "London", np.nan, "Paris"]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Drop rows with any NaN values
cleaned = df.dropna()
print("\nAfter dropna():")
print(cleaned)

# You can drop all columns with missing values: df.dropna(axis=1)
# Drop only if all values are NaN:df.dropna(how="all")
# Set a threshold (keep rows with at least 2 non‑NaN values): df.dropna(thresh=2)
```

```python
# Create a DataFrame from lists (with some NaN values)
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, np.nan, 35, np.nan],
    "City": ["New York", "London", np.nan, "Paris"]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Fill missing values
# Example 1: Fill NaN in Age with mean
df_filled_age = df.copy()
df_filled_age["Age"] = df_filled_age["Age"].fillna(df_filled_age["Age"].mean())

# Example 2: Fill NaN in City with "Unknown"
df_filled_city = df.copy()
df_filled_city["City"] = df_filled_city["City"].fillna("Unknown")

print("\nAfter fillna() on Age:")
print(df_filled_age)

print("\nAfter fillna() on City:")
print(df_filled_city)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David']
age = [24, None, 18, 47]
salary = [50000, 80000, None, 120000]

df = pd.DataFrame({"Names": names, "Age": age, "Salary": salary})

#Fill missing values in 'Age' using the backward fill
df['Age'] = df['Age'].bfill()
#Fill missing values in 'Salary' using the forward fill
df['Salary'] = df['Salary'].ffill()
df
```

Sorting data
- sort_values(): Used to sorrt data in a DataFrame or a Series in ascending or descending order based on one or more column. The by parameter is used to specify the column(s) to sort by.
- sort_index(): sort based on index.The ascending parameter is used as a boolean.
- nsmallest() and nlargest()



```python
# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 30, 35, 40, 28],
    "Score": [85, 90, 95, 88, 76]
}

df = pd.DataFrame(data, index=["e", "d", "c", "b", "a"])
print("Original DataFrame:\n", df)

# -------------------------------------------------
# 1. sort_values() - Sort by column(s)
# -------------------------------------------------
# Sort by Score ascending
df_sorted_score = df.sort_values(by="Score")
print("\nSorted by Score (ascending):\n", df_sorted_score)

# Sort by Age descending
df_sorted_age = df.sort_values(by="Age", ascending=False)
print("\nSorted by Age (descending):\n", df_sorted_age)

# -------------------------------------------------
# 2. sort_index() - Sort by row index
# -------------------------------------------------
# Sort by index ascending
df_index_asc = df.sort_index(ascending=True)
print("\nSorted by Index (ascending):\n", df_index_asc)

# Sort by index descending
df_index_desc = df.sort_index(ascending=False)
print("\nSorted by Index (descending):\n", df_index_desc)

# -------------------------------------------------
# 3. nsmallest() - Get smallest n values
# -------------------------------------------------
lowest_scores = df.nsmallest(2, "Score")
print("\n2 Smallest Scores:\n", lowest_scores)

# -------------------------------------------------
# 4. nlargest() - Get largest n values
# -------------------------------------------------
highest_scores = df.nlargest(2, "Score")
print("\n2 Largest Scores:\n", highest_scores)
```

```python
products = ['Cars', 'Bags', 'Phones', 'Shoes',
            'computers', 'clothes', 'Books', 'Boats',
            'Car oil', 'Scooter']

categories = ['Transportation', 'Household', 'Electronics',
              'Personal Items', 'Electronics', 'Personal Items',
              'Personal Items', 'Transportation', 'Transportation', 'Transportation']

sales = [2000, 1500, 1000, 1020, 2000, 1000, 1000, 1300, 1000, 2500]

costs = [1500, 1000, 700, 800, 5500, 700, 1000, 900, 1000, 1100]

profit = [500, 500, 300, 220, 500, 300, 500, 300, 100, 1400]

# Creating a DataFrame from the four lists above
df = pd.DataFrame({
    "Products": products,
    "Categories": categories,
    "Sales": sales,
    "Costs": costs,
    "Profit": profit
})

df.head()
```

```python
#histplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Sales'], kde=True)
plt.title('Sales Values distribution')
plt.show()
```

```python
#lineplot()
#VVisualize the relationship between two continuous variables
#Help identify trends,patterns, and potential correlations.

sns.lineplot(x=df['Sales'], y=df['Costs'], data=df)
plt.title('Sales and Costs analysis')
plt.show()
```

```python
#pairplot
#visualise pairwise relationships in a dataset.

sns.pairplot(df, hue="Categories")
plt.title("Pair plot of products", y=3.10, x=-1, fontsize = 20)
plt.show()
```

```python
#regplot
# Used to visualise the relationship between two numeric variables and fit a linear regression model in the data.
# Combines scatterplot and regression line in a single plot

sns.regplot(x=df["Sales"], y=df["Profit"], data=df)
plt.title("Sales and Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
