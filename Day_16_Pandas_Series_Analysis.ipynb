Using pandas, create a pandas Series from the list below.
Write a code to find how many times each item appears in
the list.

```python
import pandas as pd
list1 = ["wood", "red", "red", "white", "blue", "red"]
# Create a pandas Series
colour_series = pd.Series(list1)
# find how many times an item appears in the list
colour_series.value_counts()
```

Using pandas, write code to check if red, white, or black
are in the Series.

```python
colors = ["red", "white", "black"]
is_in_series = colour_series.isin(colors)
is_in_series
```

Using pandas, write code to return all unique values from
the Series.

```python
unique_values = colour_series.unique()
unique_values
```

Using the pandas update() method, write code to update the
Series you created in question 1. Add green and orange to
the series. The color green will be at index 0, and orange
will be at index 2.

```python
updates = {0: "green", 2: "orange"}
colour_series.update(pd.Series(updates))
colour_series
```

Write code to check the number of dimensions of your
updated Series (question 4). Write another line of code to
convert the Series into a DataFrame. Your DataFrame
should have one column of colors. Check the shape and
number of dimensions of your DataFrame. Your
DataFrame must have a shape of (6, 1).

```python
# Checking for dimensions in the series
colour_series.ndim

# Update the series to a DataFrame
colour_df = pd.DataFrame(colour_series, columns=["colors"])
colour_df

# Checking the dimensions for the DataFrame
colour_df.ndim

# Checking the shape of the DataFrame
colour_df.shape
```
