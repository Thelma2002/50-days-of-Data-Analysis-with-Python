```python
import numpy as np
list_numbers = [[12, 23, -45], [18, -77, -44]]
```

Using NumPy, create an array from the list. Write a code to
return all the negative numbers in the list.

```python
# Convert to NumPy array
arr = np.array(list_numbers)

# Boolean indexing to filter negative numbers
negative_numbers = arr[arr < 0]

print("Negative numbers:", negative_numbers)
```

Using NumPy, write code to multiply each element in the

```python
new_array = np.multiply(arr, 2)
new_array
```

 Write code to calculate the sum of the numbers on axis-1 of
the array you created in question 2.

```python
# Calculate sum along axis=1 (row-wise)
row_sum = np.sum(new_array, axis=1)
row_sum
```

Write code to calculate the sum of numbers on axis-0 of the

```python
#Calculate sum along axis=0 (column-wise)
row_column = np.sum(new_array, axis=0)
row_column
```
