```python
import numpy as np
nums = [12, 4, 6, 7, 9, 21, 67, 8]
```

1. Create a NumPy one-dimensional array with the list above.



```python
array = np.array(nums)
array
```

2. Using NumPy, write code that returns the minimum
number in the array.

```python
min = np.min(array)
print(min)
```

3. Write a code that returns the index of the largest number
in the array.

```python
max_index = np.argmax(array)
print(max_index)
```

4. Using NumPy, calculate the average of the biggest and
smallest numbers in the array.

```python
min_num = np.min(array)
max_num = np.max(array)
avg = np.array([min_num, max_num]).mean()
print(avg)
```

5. Write a code to find the mean, median, and standard
deviation of the array.


```python
mean_num = np.mean(array)
median_num = np.median(array)
std_num = np.std(array)
print(f"The mean is {mean_num}, The median is {median_num}, The standard deviation is {std_num}")
```

6. Write code to find the outlier in the array above.

```python
import numpy as np
# First sort the array
sorted_array = np.sort(array)

def calculate_outlier(num_array):
  q1 = np.percentile(num_array, 25)
  q3 = np.percentile(num_array, 75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = [x for x in num_array if x < lower_bound or x > upper_bound]
  return outliers

print(calculate_outlier(sorted_array))
```
