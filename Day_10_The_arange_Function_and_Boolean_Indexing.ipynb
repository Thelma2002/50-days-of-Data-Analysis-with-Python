Using NumPy, write a code that returns all the numbers in

Using Boolean indexing of NumPy arrays, return the days
of the week where hours worked were 8.

```python
days = ["mon", "tue",  "wed",  "thu", "Fri"]
hours_worked = [ 8, 8, 10, 11, 7]

# Creating arrays
days_arr = np.array(days)
hours_arr = np.array(hours_worked)

# Using boolean to get the 8 hour days
print(days_arr[hours_arr == 8])
```

Write another code (using Boolean indexing) to return the
day of the week with the highest number of hours worked.

```python
# Finding max value index
max_hours_index = np.argmax(hours_arr)
print(days_arr[max_hours_index])
```

Write another code to return all the days of the week where
over 9 hours of work were achieved.

```python
days_over_9hrs = days_arr[hours_arr > 9]
print(days_over_9hrs)
```
