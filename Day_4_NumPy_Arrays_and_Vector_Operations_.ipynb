1. Create two arrays of random integers between 10 and 20.
The shape of the array must be (2, 3) for the first array
and (1, 3) for the second array. Check the shape of the
arrays. Ensure that the results are reproducible.

```python
import numpy as np
#Create a random number generator with a fixed seed
rng = np.random.default_rng(seed=42)
#Generate the arrays
arr1 = rng.integers(low=10, high=21, size=(2,3))
arr2 = rng.integers(low=10, high=21, size=(1,3))
#Check the shapes
print("Shape of array 1:", arr1.shape)
print("Shape of array 2:", arr2.shape)
```

2. Write a code to add the two arrays you just created in
question 1. Check the shape of the resulting array. Explain
why the resulting shape is that of the bigger array from the
two arrays created in question 1.

```python
arr3 = np.add(arr1, arr2)
print("Array 1",arr1)
print("Array 2",arr2)
print("Array 3",arr3)
print("Shape of array 3:", arr3.shape)
#Under the broadcasting rule the smaller array is expanded to match the shape of the larger array
```

3. Write another code to perform a dot operation on the
arrays.

The dot product (or matrix multiplication) combines two arrays in a way that multiplies rows by columns.

- For 1D arrays (vectors):
The dot product is the sum of element‑wise multiplications.
Example:
[1,2,3] . [4,5,6]= 1 . 4 + 2 . 5+ 3 . 6 = 32
- For 2D arrays (matrices):
The dot product is matrix multiplication.
Rule: The number of columns in the first array must equal the number of rows in the second array.
Example:
If A is shape (2,3) and B is shape (3,4), then A.dot(B) will be shape (2,4).



```python
# To perform a dot product, we need to align dimensions. Let’s transpose arr2 to shape (3,1) so the multiplication works.
result = np.dot(arr1, arr2.transpose())   #  This line can also be result = np.dot(arr1,arr2.T)
print(result)
print(result.shape)
```

4. Using NumPy, create a 1-dimensional array of 100
random integers from 0 to 10. Use the array to create a
histogram of the data and calculate the median and mode.
Ensure that the results are reproducible.

```python
import matplotlib.pyplot as plt
# Step 1: Create a random number generator with a fixed seed
rng = np.random.default_rng(seed=42)

# Step 2: Generate a 1D array of 100 random integers between 0 and 10
arr = rng.integers(low=0, high=11, size=100)

# Step 3: Create a histogram from the data
plt.hist(arr, bins=11, edgecolor="red", )
plt.title("Histogram of random integers")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xticks(range(0, 11))
plt.figure(figsize=(12,8))
plt.show()
```

```python
#Calculate the median and mode
from statistics import mode
# from scipy import stats   # another one for mode
median_value = np.median(arr)
print("Median is:", median_value)

mode_value = mode(arr)
print("Mode is:", mode_value)
# mode_val = stats.mode(arr, keepdims=True)  #For the secodn method of importing
```

5. Using NumPy, create a 3-dimensional array of 100 random
floats between 0 and 1. Use the array to create a 3D scatter
plot of the data. Ensure that your results are reproducible.

```python
# Generate values
rng = np.random.default_rng(seed=42)
#Generate random floats between 0 and 1
values = rng.random((100, 3)) #100 points in 3d space
# Plot scatter plot
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection="3d")

ax.scatter(values[:,0], values[:,1], values[:,2], c="purple", marker="o")
plt.title("3D Scatter plot")
plt.show()
```

6. Create a an array from the list below:

lst =[[0, 1, 3, 0, 4], [8, 9, 0, 9, 6]]

Write code to flatten the list and return all non-zero
numbers from the list.

```python
lst =[[0, 1, 3, 0, 4], [8, 9, 0, 9, 6]]
#Create a numpy array
arr = np.array(lst)
# Flatten the array
flat_arr = arr.flatten()
#filter non-zero numbers
non_zero = flat_arr[flat_arr != 0]

print("Flattened array:", flat_arr)
print("Non-Zero array:", non_zero)
```
