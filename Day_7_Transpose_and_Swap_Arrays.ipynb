 Using the arange() function, create an array of numbers
from 0 and 10. The shape of the array must be (2, 5).

```python
import numpy as np
array = np.arange(0, 10).reshape(2, 5)
array
```

Write a code to change the shape of the array (question 1)
to (5, 2). Create a new variable for this array. Now, add this
array to the original array in question 1. The resulting array
should have shapes (2, 5). Save this as a variable.

```python
# reshaping the array
array2 = array.reshape(5,2)
array2

# adding array to origional array
array3 = array2.transpose() + array
array3
```

Swap the indexes of the array (question 2) using the
swapaxis() method. Create a new variable for the swapped
array.

```python
swapped_array = np.swapaxes(array3, 0, 1)
swapped_array
```

Write a code to slice row 1 from the swapped array in
question 3.

```python
row_one = swapped_array[0, :]
row_one
```
