Using the data below, create an array and transpose it. The
creation of an array and transposition should be combined
into one code. The code should return a transposed array.

```python
import numpy as np
names = ["John", "Kelly", "Jos", "Peter", "Robert", "Piper"]
age = [21, 21, 56, 44, 56, 96]
gender = ['m', 'f', 'm', 'm', 'm', 'f']

# Create an array and transpose it
arr = np.array([names, age, gender]).T
print(arr)
```

Using slicing or indexing, write code to return the last row
of the transposed array (question 1).

```python
# Slicing or indexing to return last row
print(arr[-1])
```

Using slicing or indexing, write code to return the first two
rows of the transposed array (question 1).

```python
# Slicing or indexing to return first two rows
print(arr[:2])
```

One of the uses of slicing is for data visualization. Slicing is
often used in data visualization to extract specific portions
of a dataset for plotting. Write code to create a subarray of
the ages (only ages) in the transposed array using slicing.
Using Matplotlib, plot a histogram of the age array. Your
graph should have an xlabel, an ylabel, and a title.

```python
import matplotlib.pyplot as plt
# slicing to return subarray of ages only
ages = arr[:, 1]

# Plot a histogram for ages showing xlabel, ylabel and title
plt.hist(ages)
plt.xLabel('Ages')
plt.yLabel('Frequency')
plt.title('Histogram of ages')
plt.show()
```
