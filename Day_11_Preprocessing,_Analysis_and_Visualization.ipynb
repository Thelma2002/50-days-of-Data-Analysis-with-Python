Using NumPy, write a
code to check how many non-numeric elements are in the
list.

```python
import numpy as np
list1 = ["34", "name",  "45",  "is",  "100"]
# Check how many non_numerical elements are in the list
non_numeric_count = 0
for item in list1:
    try:
        int(item) # Attempt to convert to integer
    except ValueError:
        non_numeric_count += 1 # If conversion fails, it's non-numeric
print(f"Number of non-numeric elements: {non_numeric_count}")
```

Using NumPy, write code to sum all the numeric elements
in the list.

```python
# sum all the numeric elements in the list
numeric_sum = sum(int(item) for item in list1 if item.isdigit())
print(f"Sum of numeric elements: {numeric_sum}")
```

Using NumPy, write code to square all the numeric
elements in the array you created in question 1. Return an
array of squared elements.

```python
# square all the numeric elements in the array
numeric_square = np.array([int(item)**2 for item in list1 if item.isdigit()])
print(f"Squared numeric elements: {numeric_square}")
```
