```python
import pandas as pd
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
miles_run = [120, 80, 100, 90, 85]
time_in_hrs = [40, 38, 45, 50, 50]
```

1. Create a DataFrame from the lists above using pandas

```python
df = pd.DataFrame({"names": names, "miles_run": miles_run, "time_in_hours": time_in_hrs})
df
```

2. More information has become available below. Create a
new DataFrame with this information. Merge it to the
DataFrame above using pandas concat() function.

```python
# new information
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
Age = [45, 28, 21, 55, 62]
gender = ["M", "M", "F", "F", "M"]

# creating new dataframe
new_df = pd.DataFrame({"names": names, "Age": Age, "gender": gender})

# Concatenate new DataFrame to Old DataFrame and drop name column
df = pd.concat([df, new_df.drop(columns="names")], axis=1)
df
```

3. Shuffling DataFrames can randomize the data. Write code
to reshuffle the index of the DataFrame.

```python
df = df.sample(frac=1)
df
```

4. Explain why it may be important to reshuffle the index of
the data for machine learning.
- In many machine learning algorithms, the order of the data
in the training set can affect the final model. For example,
if the data is ordered by the target variable, an algorithm
that builds a model incrementally, like gradient descent,
will have an easier time fitting the model to the data.

- Shuffling the data can help prevent overfitting. If the data
is ordered by the target variable, an algorithm that uses all
the data to make predictions will tend to overfit the data
because it will learn to predict the target variable based on
the order of the data rather than the underlying patterns in
the data.

5. Using the pandas melt() function, reshape your DataFrame
to a long format. The "names" column must remain in
place.

```python
df_set_long = pd.melt(df, id_vars=["names"],
                      var_name="Variable",
                      value_name="Values")
df_set_long
```
