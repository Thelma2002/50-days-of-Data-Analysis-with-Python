Using NumPy, write code that returns all the names and
the  number of  students from class "e" who got over 50
marks.

```python
import numpy as np
names = ["Kelly", "Ben", "Jack", "Muhammad", "Jose", "Liz"]
grades = [45, 65, 75, 35, 85, 40]
classes = ['a', 'c', 'e', 'd', 'e', 'f']

# Create the students array. NumPy will coerce all elements to string if mixed types are present in rows.
student_arr = np.array([names, [str(g) for g in grades], classes])
print(student_arr)

# Convert the grades row to integer type for numerical comparison
grades_numeric = student_arr[1].astype(int)

# Create boolean masks for filtering students (columns)
class_e_condition = (student_arr[2] == 'e')
over_50_condition = (grades_numeric > 50)

# Combine the conditions using logical AND to get a mask for students that meet both criteria
combined_condition = class_e_condition & over_50_condition

# Apply the combined condition to select the relevant columns from student_arr
filtered_students_data = student_arr[:, combined_condition]

# Extract names and count the number of students
filtered_names = filtered_students_data[0]
number_of_students = len(filtered_names)

print(f"Names of students from class 'e' who got over 50 marks: {filtered_names}")
print(f"Number of students from class 'e' who got over 50 marks: {number_of_students}")
```

Your next task is to write a code that returns all the students
who got less than 75 marks.

```python
# Return all students who got less than 75 marks
less_than_75_condition = (grades_numeric < 75)
less_than_75_students = student_arr[:, less_than_75_condition]
print(f"Students who got less than 75 marks: {less_than_75_students}")
```

Write code that returns how many marks a student named
Liz got.

```python
# How many marks did liz get
liz_index = np.where(student_arr[0] == 'Liz')[0]
if len(liz_index) > 0:
    liz_marks = student_arr[1][liz_index]
    print(f"Liz got {liz_marks} marks.")
else:
    print("Liz not found in the array.")
```

 Which student got the highest marks?

```python
# Return student with the highest mark
highest_mark_index = np.argmax(grades_numeric)
highest_mark_student = student_arr[:, highest_mark_index]
print(f"Student with the highest mark: {highest_mark_student}")
```

What is the longest name in the array, and what is its
index?  

```python
# Return student witht the longest name and it's index
student_names = student_arr[0]
longest_name_index = np.argmax([len(name) for name in student_names])
longest_name_student = student_arr[:, longest_name_index]
print(f"Student with the longest name: {longest_name_student}")
print(f"Index of the student with the longest name: {longest_name_index}")
```
